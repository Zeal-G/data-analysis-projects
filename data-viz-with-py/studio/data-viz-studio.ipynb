Below we set up the dataframe with the Goodreads dataset for you. If you run this cell, you may get an error that some lines were skipped. Do not worry about that! For this studio, we want to focus on creating an effective visualization.

```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
df.head(20)
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
title = ["Harry Potter and the Order of the Phoenix", "Neither Here nor There: Travels in Europe", "The Hitchhiker's Guide to the Galaxy"]
average_rating = [4.49, 3.86, 4.22]

plt.bar(title, average_rating)
```

```python
# Use your Python skills to create a visualization here.
data = [["Harry Potter and the Chamber of Secrets", 4.42], ["The Hitchhiker's Guide to the Galaxy", 4.22], ["Notes from a Small Island", 3.91]]
books = pd.DataFrame(data, columns=["title","average_rating"])

sns.barplot(books, x="average_rating", y="title", orient="y")
```
